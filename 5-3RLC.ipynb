```python
x1,x2 = var('x1 x2')
f1 = -x1^3 - 4*x1 - x2
f2 = 3*x1
```

```python
f11 = diff(f1,x1);show(f11)
f12 = diff(f1,x2);show(f12)
f21 = diff(f2,x1);show(f21)
f22 = diff(f2,x2);show(f22)
```

```python
C1 = implicit_plot(f1,(x1,-2,2),(x2,-9,9),rgbcolor='blue')
C2 = implicit_plot(f2,(x1,-2,2),(x2,-9,9),rgbcolor='red')
VF1 = plot_vector_field([f1,f2],(x1,-2,2),(x2,-9,9))
show(C1+C2+VF1,aspect_ratio = 2/10,gridlines = true)
#show(C1+C2)
```

```python
C12 = implicit_plot(f1,(x1,-1,1),(x2,-3,3),rgbcolor='blue')
C22= implicit_plot(f2,(x1,-1,1),(x2,-3,3),rgbcolor='red')
VF12 = plot_vector_field([f1,f2],(x1,-1,1),(x2,-3,3))
show(C12+C22+VF12,aspect_ratio = 2/10,gridlines = true)
#show(C1+C2)
```

```python
M = Matrix([[f11,f12],[f21,f22]]).subs(x1=0,x2=0);show(M)
```

```python
Eigens = M.eigenvalues();show(Eigens)#<0, therefore stable
```
