```python
t = var('t')
```

```python
#New static state values
x10 = (10000-6000*t) / (1-t^2)
x20 = (6000-10000*t) / (1-t^2)
```

```python
f1s(x1,x2) = .10*x1 - .10*x1^2/10000 - .10*t*x1*x2/10000
f2s(x1,x2) = .25*x2 - .25*x2^2/6000 - .25*t*x1*x2/6000
```

```python
f11s = diff(f1s,x1)
f12s = diff(f1s,x2)
f21s = diff(f2s,x1)
f22s = diff(f2s,x2)
show([[f11s,f12s],[f21s,f22s]])
```

```python
Ms = Matrix([[f11s(x10,x20),f12s(x10,x20)],[f21s(x10,x20),f22s(x10,x20)]])
show(Ms)
Eigenss = Ms.eigenvalues()
show(Eigenss)
print(Eigenss)
```

```python
N = 600
tn = [None] * N
lambda1 = [None] * N
lambda2 = [None] * N
for i in range(0,N-1):
 tn[i] = i/1000
 lambda1[i] = float(Eigenss[0].subs(t=tn[i]))
 lambda2[i] = float(Eigenss[1].subs(t=tn[i]))
```
